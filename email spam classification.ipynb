```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("mail_data.csv")
```

```python
data = df.where((pd.notnull(df)), '')
```

```python
data.shape
```

```python
data.loc[data['Category']=='spam', 'Category', ] = 0
data.loc[data['Category']=='ham', 'Category', ] = 1
```

```python
X = data['Message']
```

```python
Y = data['Category']
```

```python
vectors = TfidfVectorizer(min_df=1, stop_words = 'english', lowercase = True)
X_features = vectors.fit_transform(X)
```

```python
Y_features = Y.astype('int')
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_features, Y_features, test_size=0.2, random_state=3)
```

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
prediction = model.predict(X_train)
```

```python
accuracy_on_training_data = format(accuracy_score(Y_train, prediction)*100, ".2f")
```

```python
print(f"Accuracy on the training data: {accuracy_on_training_data}%")
```

```python
ultimate_prediction = model.predict(X_test)
prediction_on_test_data = format(accuracy_score(Y_test, ultimate_prediction)*100, '.2f')
```

```python
print(f"Accuracy on the testing data: {prediction_on_test_data}%")
```
